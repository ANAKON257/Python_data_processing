# Handle the missing values.

```python
#Import the packages
import pandas as mypd
```

```python
#import the data
mydata=mypd.read_excel(r'D:\Python_ISI\data\Preprocessing_Data_I.xlsx')
```

```python
#Explore the data
mydata.head()
```

```python
#check for missing values
mydata.info()
```

```python
#Deleting the records with missing values
mynewdata=mydata.dropna()
```

```python
mynewdata
```

```python
mynewdata.info()
```

```python
mynewdata.to_excel(r'D:\Python_ISI\Cleaned_data.xlsx')
```

```python
#Replace missing values with feature means
mydatanew=mydata.fillna(mydata.mean())
```

```python
mydatanew
```

```python
mydatanew.to_excel(r'D:\Python_ISI\Cleaned_data.xlsx')
```

```python
#Replace missing values in Review time with mean
myrevtime=mydata.Review_Time
```

```python
myrevtime.fillna(myrevtime.mean(),inplace=True)
```

```python
mydata
```

```python
#Replace missing values in Test Coverage with median
mytestcov=mydata.Test_Coverage
```

```python
mytestcov.fillna(mytestcov.median(),inplace=True)
```

```python
mydata
```

```python
#Replace missing values in Test Coverage with 95
mydata.Review_Coverage.fillna(95,inplace=True)
```
