```python
#Import the packages
import pandas as mypd
import matplotlib.pyplot as myplot
```

```python
#Import the data
mydata=mypd.read_excel('D:/Python_ISI/data/Productivity.xlsx')
```

```python
mydata.head()
```

```python
#copy first 20 records to a new data table
mynewdata=mydata.iloc[0:20,:]
mynewdata
```

```python
mynewdata.to_excel(r'D:\Python_ISI\Newdata.xlsx')
```

```python
#!pip3 install jedi==0.17.2
```

```python
#copy the fields to new variable
myprod=mydata.Productivity
myexp=mydata.Developer_Experience
myreuse=mydata.Code_Reuse
mykrusage=mydata.Knowledge_Repository_Usage
```

```python
myprod.describe()
```

```python
#Average Proudtivity with experience
myprod.groupby(myreuse).mean()
```

```python
#Boxplot  of productivity with experience
mydata.boxplot(column=['Productivity'],by=['Developer_Experience'],grid=False)
myplot.show()
```

```python
#Average Productivity with Code Reuse
myprod.groupby(myreuse).mean()
```

```python
mydata.boxplot(column=['Productivity'],by=['Code_Reuse'],grid=False)
myplot.show()
```

```python
myprod.groupby(mykrusage).mean()
```

```python
mydata.boxplot(column=['Productivity'],by=['Knowledge_Repository_Usage'],grid=False)
myplot.show()
```

```python
#Average Productivity with Experience & Code Reuse
myprod.groupby([myexp,myreuse]).mean()
```

```python
#Boxplot  of profuctivity with experience and code reuse
mydata.boxplot(column=['Productivity'],by=['Developer_Experience','Code_Reuse'],grid=False)
myplot.show()
```

```python
#Average Productivity with Experience , Code Reuse & K R Usage
myprod.groupby([myexp,myreuse,mykrusage]).mean()
```

```python
#Boxplot  of profuctivity with experience , code reuse & K R Usage
mydata.boxplot(column=['Productivity'],by=['Developer_Experience','Code_Reuse','Knowledge_Repository_Usage'],grid=False)
myplot.show()
```

```python
#Median Productivity with Experience , Code Reuse & K R Usage
myprod.groupby([myexp,myreuse,mykrusage]).median()
```

```python
#Median Productivity with Experience , Code Reuse & K R Usage
myresult=myprod.groupby([myexp,myreuse,mykrusage]).describe()
```
