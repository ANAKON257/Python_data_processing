# FEATURE SELECTION: Recursive Feature Elimination Method - Python Example:-
Select the 5 important features in the Boston housing data. The
response variable is MEDV

```python
#Import packages
import pandas as mypd
from sklearn.feature_selection import RFE
from sklearn import tree
from sklearn.ensemble import RandomForestRegressor
```

```python
#Import the data
mydata=mypd.read_excel('D:/Python_ISI/data/Boston_Housing Data.xls')
```

```python
#Explore the data
mydata.head()
```

```python
#Separate x's & y
x=mydata.iloc[:,0:13]
y=mydata.MEDV
```

```python
x.head()
```

```python
#Specify the model
#mymodel=tree.DecisionTreeRegressor() #for drcision tree
mymodel=RandomForestRegressor()
```

```python
#Set up feature selection
mysearch=RFE(mymodel,n_features_to_select=1,step=1)
```

```python
#Feature Selection
mysearch=mysearch.fit(x,y)
```

```python
#Top 5 features
mydecision=mysearch.support_
```

```python
mydecision=mypd.DataFrame(mydecision,columns=['Decision'])
```

```python
mydecision
```

```python
#Featuree names
myfeatures=x.columns
```

```python
myfeatures=mypd.DataFrame(myfeatures,columns=['Features'])
```

```python
#Important features
myresult=myfeatures.join(mydecision)
```

```python
myresult
```

```python
#Ranking of the features
myranks=mysearch.ranking_
```

```python
myranks=mypd.DataFrame(myranks,columns=['Ranks'])
```

```python
myresult=myresult.join(myranks)
```
