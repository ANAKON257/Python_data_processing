# Recursive Feature Elimination Method Example:
Identify 5 most important features deciding whether a customer will take
pep or not in bank-data?

```python
#Import packages
import pandas as mypd
from sklearn.feature_selection import RFE
from sklearn import tree
from sklearn.ensemble import RandomForestRegressor
```

```python
#Import the data
mydata=mypd.read_csv('D:/Python_ISI/data/bank-data.csv')
```

```python
mydata.head()
```

```python
#Separate x's & y
x=mydata.iloc[:,0:10]
y=mydata.pep
```

```python
#Specify the model
mymodel=tree.DecisionTreeClassifier()
```

```python
#Set up feature selection
mysearch=RFE(mymodel,n_features_to_select=5,step=1)
```

```python
#Feature Selection
mysearch=mysearch.fit(x,y)
```

```python
#Top 5 features
mydecision=mysearch.support_
```

```python
mydecision=mypd.DataFrame(mydecision,columns=['Decision'])
```

```python
mydecision
```

```python
#Featuree names
myfeatures=x.columns
```

```python
myfeatures=mypd.DataFrame(myfeatures,columns=['Features'])
```

```python
#Important features
myresult=myfeatures.join(mydecision)
```

```python
myresult
```

```python
#Ranking of the features
myranks=mysearch.ranking_
```

```python
myranks=mypd.DataFrame(myranks,columns=['Ranks'])
```

```python
myresult=myresult.join(myranks)
```
