# ADVANCED IMPUTATION METHODS: KNN Imputation using Python - Example
The sprint productivity data is given in Preprocessing_Data_I and
Preprocessing_Data_II files. Kindly append the files and replace the missing
values using k-nn (k = 4)

```python
#Import the packages
import pandas as mypd
from sklearn.impute import KNNImputer
```

```python
#Import the data
mydata1=mypd.read_excel('D:/Python_ISI/data/Preprocessing_Data_I.xlsx')
```

```python
#Explore the data
mydata1.head()
```

```python
#Check for missing values
mydata1.info()
```

```python
#Import the 2nd data
mydata2=mypd.read_csv('D:/Python_ISI/data/Preprocessing_Data_II.csv')
```

```python
mydata2.head()
```

```python
mydata2.info()
```

```python
#Append the files
mydata=mydata1.append(mydata2,ignore_index=True)
```

```python
mydata=mydata.iloc[:,1:10]
mydata.head()
```

```python
mydata.info()
```

```python
mydata
```

```python
#Replace missing values using KNN method(k=4)
knn=KNNImputer(n_neighbors=4)
```

```python
cleaned_data=knn.fit_transform(mydata)
```

```python
cleaned_data=mypd.DataFrame(cleaned_data,columns=mydata.columns)
```

```python
cleaned_data.info()
```

```python
cleaned_data
```

```python
cleaned_data.to_excel('D:/Python_ISI/Cleaned.xlsx')
```
