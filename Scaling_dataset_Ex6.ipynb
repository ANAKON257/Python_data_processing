# TRANSFORMATION / NORMALIZATION Exercise : 
The TAT data of a tech support process is given in TAT file. Normalize the variables in the TAT data?

```python
#Import the package
import pandas as mypd
from sklearn.preprocessing import StandardScaler as z
from sklearn.preprocessing import MinMaxScaler as minmax
```

```python
#import the data
mydata=mypd.read_excel('D:/Python_ISI/data/TAT.xlsx')
```

```python
#explore the data
mydata.head()
```

```python
#Descriptive summary
mydata.describe()
```

```python
#z transform
myzdata=z().fit_transform(mydata)
```

```python
#change the format to dataframe
myzdata=mypd.DataFrame(myzdata,columns=mydata.columns)
```

```python
myzdata.head()
```

```python
#descriptive summary
round(myzdata.describe(),3)
```

```python
myzdata.to_excel('D:/Python_ISI/Transformed_data.xlsx')
```

```python
#Min Max transformation
tr_data=minmax().fit_transform(mydata)
```

```python
#change the format to dataframe
tr_data=mypd.DataFrame(tr_data,columns=mydata.columns)
```

```python
#Descriptive summary
tr_data.head()
```

```python
round(tr_data.describe(),3)
```
